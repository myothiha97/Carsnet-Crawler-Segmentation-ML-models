```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df1 = pd.read_csv("../datas/imported_cars_2020-11-05 09_42_51.csv",na_values=['-'])
df2 = pd.read_csv('../datas/imported_cars_2020-11-06 11_45_43.csv',na_values=['-'])
df = pd.concat([df1,df2])
df
```

```python
# df['Mileage'].replace('-',np.NaN,inplace=True)
df.dropna(subset = ["Mileage"], inplace=True)
```

```python
df
```

```python
df['Mileage']= df["Mileage"].str.replace("[*]",'0')
```

```python
df['Mileage']
```

```python
df[df.Mileage.str.contains('-')]
```

```python
df['Mileage'] = df.Mileage.str.replace(r'[a-zA-Z]|\s|\-|[,]','')
df = df[df['Mileage'] != '']
```

```python
df.Mileage
```

```python
df = df[df['Mileage'] != '0152760']
```

```python
df
```

```python
df = df[df.Mileage.map(len) < 7]
```

```python
df
```

```python
y = df.Mileage.astype('int64')
```

```python
y
```

```python
x = df["Post Detail"]
```

```python
ascii_x = []
for post in x:
    lines = post.split('\n')
    ascii_char= []
    for line in lines:
        for word in line:
            for char in word:
#                 ascii_x.append(ord(char.lower())/1000)
                ascii_char.append(ord(char.lower()))
#                 print(char,'a')
    ascii_x.append(ascii_char)
```

```python
df_x = pd.DataFrame(ascii_x)
```

```python
df_x
```

```python
df_x.fillna(0,inplace=True)
```

```python
df_x
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainx,testx,trainy,testy = train_test_split(df_x,y,test_size=0.2,random_state = 3)
```

```python
from xgboost import XGBRegressor
```

```python
xgb = XGBRegressor(n_estimators=200,n_jobs= -1)
xgb.get_params()
```

```python
train_gbr = xgb.fit(trainx,trainy)
```

```python
train_gbr.score(testx,testy)
```

```python
import pickle
pickle.dump(train_gbr,open("XGboost_Mileage_5%.pkl",'wb'))
```

```python
errors = []
for i in range(200):
#     predx = pd.Series.reshape(testx.iloc[i])
#     predx.reshape(3684,1)
#     predx.res
    predy = train_gbr.predict(testx.iloc[[i]])
    print(round(predy[0],2),'-------------',testy.iloc[i])
    errors.append(abs(predy[0]-testy.iloc[i]))
```
