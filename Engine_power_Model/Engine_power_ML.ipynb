```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv("../datas/imported_cars_2020-11-05 09_42_51.csv")
df.T
```

```python
df['Engine'].replace('-',np.NaN,inplace=True)
df.dropna(subset = ["Engine"], inplace=True)
# df = df[df['Brand Name'].map(len) <= 13]
```

```python
df
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df["Engine"]=df["Engine"].str.replace(r"[a-zA-Z]+",'')
df = df[df['Engine'].map(len) > 3]
df["Engine"] = df["Engine"].str.replace(r"\s+",'')
df["Engine"]
```

```python
y = df.Engine.astype('int')
y
```

```python
x = df["Post Detail"]
```

```python
df['Engine'].value_counts().plot(kind='bar',figsize=(30, 20),color='b')
plt.xticks(fontsize=20);
plt.yticks(fontsize=20);
```

```python
ascii_x = []
for post in x:
    lines = post.split('\n')
    ascii_char= []
    for line in lines:
        for word in line:
            for char in word:
#                 ascii_x.append(ord(char.lower())/1000)
                ascii_char.append(ord(char.lower()))
#                 print(char,'a')
    ascii_x.append(ascii_char)
```

```python
df_x = pd.DataFrame(ascii_x)
```

```python
df_x
```

```python
df_x.fillna(0,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainx,testx,trainy,testy = train_test_split(df_x,y,test_size=0.2,random_state = 3)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rdf = RandomForestRegressor(n_estimators=200,n_jobs= -1)
```

```python
rdf.fit(trainx,trainy)
```

```python
rdf.score(testx,testy)
```

```python
predy =rdf.predict([testx.iloc[7]])
```

```python
load_rdf = pickle.load(open('Random_forest_Engine_model1.pkl','rb'))
```

```python
errors = []
for i in range(200):
    predy = load_rdf.predict([testx.iloc[i]])
    ans = "{:.2f}".format(predy[0])
    print(ans,"-----------",testy.iloc[i])
    error = abs(predy[0]-testy.iloc[i])
    errors.append(error)
```

```python
avg_error = np.mean(errors)
```
