```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv("../datas/imported_cars_2020-11-05 09_42_51.csv")
```

```python
df['Grade'].replace('-',np.NaN,inplace=True)
df.dropna(subset = ["Grade"], inplace=True)
```

```python
df
```

```python
df = df[df['Grade'] != ".."]
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df["Grade"] = df["Grade"].str.replace(r"\d+|[{}/><~!@#$%%^&*]+",'')
```

```python
df_y = pd.Categorical(df['Grade']).codes
grade_codes = {}
for i , val in enumerate(df['Grade'].items()):
    grade_codes[val[1]] = df_y[i]
#     print(val)
# color_codes
import csv
with open("Grade_Codes.csv",'w') as file:
    columns = ['Colors',"Codes"]
    writer = csv.DictWriter(file,grade_codes.keys())
    writer.writeheader()
    writer.writerow(grade_codes)
```

```python
grade_codes
```

```python
x = df["Post Detail"]
```

```python
ascii_x = []
for post in x:
    lines = post.split('\n')
    ascii_char= []
    for line in lines:
        for word in line:
            for char in word:
#                 ascii_x.append(ord(char.lower())/1000)
                ascii_char.append(ord(char.lower()))
#                 print(char,'a')
    ascii_x.append(ascii_char)
```

```python
df_x = pd.DataFrame(ascii_x)
```

```python
df_x
```

```python
df_x.fillna(0,inplace=True)
```

```python
df_x
```

```python
from sklearn.model_selection import train_test_split
trainx,testx,trainy,testy = train_test_split(df_x,df_y,test_size=0.2,random_state = 3)
from sklearn.ensemble import RandomForestClassifier
rdf = RandomForestClassifier(n_estimators=200,n_jobs= -1)
```

```python
train_rdf = rdf.fit(trainx,trainy)
```

```python
train_rdf.score(testx,testy)
```

```python
import pickle
pickle.dump(train_rdf,open('Random_Forest_Classifier_Grade_33%.pkl','wb'))
```

```python
for i in range(0,200): 
    predy=train_rdf.predict([testx.iloc[i]])
    realy = testy[i]
    for key,val in grade_codes.items():
        if predy == val:
            predy = key
        if realy == val:
            realy = key
    print(predy,'---------------',realy)
```
