```python
df = pd.read_csv("../datas/imported_cars_2020-11-11 08_48_11.csv",na_values=null_values)
```

```python
df.dropna(subset = ["Model"], inplace=True)
```

```python
df.Model.value_counts()
```

```python
df = df[df["Model"].map(len) < 9]
```

```python
df
```

```python
x = df['Post Detail']
```

```python
ascii_x = []
for post in x:
    lines = post.split('\n')
    ascii_char= []
    for line in lines:
        for word in line:
            for char in word:
#                 ascii_x.append(ord(char.lower())/1000)
                ascii_char.append(ord(char.lower()))
#                 print(char,'a')
    ascii_x.append(ascii_char)
```

```python
df_x = pd.DataFrame(ascii_x)
```

```python
df_x
```

```python
df_x.fillna(0,inplace=True)
```

```python
df_x
```

```python
y = df.Model
```

```python
ascii_y = []
for word in y:
    ascii_char= []
#     if len(word) > 13:
#         continue
    for char in word:
        ascii_char.append(ord(char.lower()))
#                 print(char,'a')
    ascii_y.append(ascii_char)
```

```python
df_y = pd.DataFrame(ascii_y)
df_y
```

```python
df_y.fillna(0,inplace=True)
df_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainx,testx,trainy,testy = train_test_split(df_x,df_y,test_size=0.2,random_state = 3)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rdf = RandomForestRegressor(n_estimators=200,n_jobs= -1)
```

```python
train_rdf = rdf.fit(trainx,trainy)
```

```python
train_rdf.score(testx,testy)
```

```python
import pickle
pickle.dump(train_rdf,open('Random_Forest_Regressor_Model_21%.pkl','wb'))
```
